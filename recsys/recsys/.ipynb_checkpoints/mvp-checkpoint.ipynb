```python
rank.columns = ['user', 'isbn', 'rank']
target_isbn = "0439136350"
target_users = rank.loc[rank.isbn == target_isbn, "user"].unique()
rank_only_users = rank[rank.user.isin(target_users)]
t = rank_only_users.groupby("isbn")["rank"].nunique() 
rated_books = t[t > 1].index
```

```python
needed_rank = rank[(rank.isbn.isin(rated_books)) & (rank.user.isin(target_users))]
target_rank = needed_rank[needed_rank.isbn == target_isbn]
target_rank.columns = ["user", "target_isbn", "target_rank"]
other_rank = needed_rank[needed_rank.isbn != target_isbn]
needed_rank = pd.merge(other_rank, target_rank, on=["user"])
```

```python
needed_rank
```

```python
stats = needed_rank.groupby(["isbn", "target_isbn"]).agg({"rank": ["mean", "std"],
                                                          "target_rank": ["mean", "std"]}).reset_index()
stats
```

```python
stats = needed_rank.groupby(["isbn", "target_isbn"]).agg({"rank": ["mean", "std"],
                                                          "target_rank": ["mean", "std"]}).reset_index()
stats.columns = ["isbn","target_isbn",  "mean", "std", "target_mean", "target_std"]
stats = stats[(stats['std'] > 0) & (stats['target_std'] > 0)]
stats.sort_values("isbn")
```

```python
t = pd.merge(needed_rank, stats, on=["isbn", "target_isbn"])
t.sort_values("isbn")
```
